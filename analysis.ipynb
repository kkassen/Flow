<h2>Module 3: Data Wrangling</h2>

```python
# Data Wrangling Object
dw = flow.dataWrangling()
```

```python
# Removing non-numeric characters from the values in the existing_loans column
dw.drop_characters(data, column='existing_loans', expression='[^0-9]+')
```

```python
# encoding the checking_status column with numeric values
dictionary = {'checking_status': {'None': 0, '<0DM': 1, '0_to_200DM': 2, '>200DM': 3}}
dw.map(data, dictionary=dictionary)
```

<h2>Module 4: Predictive Modeling</h2>

```python
# Predictive Modeling Object
pm = flow.predictiveModeling()
```

```python
# Partitioning the data into testing and training datasets
split = pm.partition(data, size=0.25)
features = ['age', 'checking_status', 'duration', 'loan_amount', 'installment_commitment']
training_data = pm.train_sets(data=split[0], features=features, column='accepted')
testing_data = pm.test_sets(data=split[1], features=features, column='accepted')
```

```python
# Reviewing the correlation table for the features
de.correlation(testing_data[0], view=False, color=True)
```

```python
# Selecting, fitting, producing predictions, and outputting performance metrics for the Generic Model
model = pm.model_selector(algorithm='tree')
pm.generic_model(M=model, features=training_data[0], target=training_data[1])
predictions = pm.predict(M=model, features=testing_data[0])
pm.measurements(testing_data[1], predictions, option='Generic',M=model, x_train=training_data[0], y_train=training_data[1])
```

```python
# Optimizing, fitting, cross-validating, producing predictions, and outputting performance metrics for the Optimized Model
hyperparameters = {'max_depth': [5]}
optimized_model = pm.optimize(M=model, hyperparmeters=hyperparameters, folds=5, metric='accuracy', features=training_data[0], target=training_data[1])
p = pm.predict(M=optimized_model, features=testing_data[0])
pm.measurements(testing_data[1], p, option='Optimized', M=optimized_model, X=data[features], Y=data['accepted'], x_train=training_data[0], y_train=training_data[1])
```

```python
# Writing the test features, actuals, and predications to a .csv file
pm.write(testing_data[0], testing_data[1], p, c1='accepted', c2='prediction', file='output.csv')
```

```python
# Visualization Object
v = flow.visualization()
```

```python
# Visualizing the Generic Model Decision Tree
v.visualize_tree(M=model, title='Generic Model')
```

```python
# Visualizing the Optimized Model Decision Tree
v.visualize_tree(M=optimized_model.best_estimator_, title='Optimized Model')
```

```python
# Visualizing the cofusion_matrix for the performance metrics
v.confusion_matrix(testing_data[1], p)
```
